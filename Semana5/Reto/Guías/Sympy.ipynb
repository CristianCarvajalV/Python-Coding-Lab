# Derivadas
Para derivar expresiones, se usa la función `diff`

```python
x, y, z, u = symbols('x y z u')
```

```python
diff(cos(x), x)
```

```python
diff(x ** 4, x, 3)
```

```python
diff(exp(x ** 2), x)
```

```python
expresion = exp(x * y * z)
expresion
```

```python
diff(expresion, x, y, z, z)
```

# Derivada no evaluada
Estos objetos no evaluados son útiles para retrasar la evaluación de la derivada o para imprimir. También se utilizan cuando SymPy no sabe cómo calcular la derivada de una expresión 

```python
# crear una derivada sin evaluarla
derivada = Derivative(expresion, x, y, z, 2)
```

```python
derivada
```

```python
# para evaluarla
derivada.doit()
```

# Integrales
Para obtener una integral, usamos la función `integrate`. Se debe tener en cuenta que hay dos tipos de integrales:
1. Indefinida o primitiva: para evaluar esta solo se necesita pasar la variable después de la expresión
2. Definida: Se pasa una tupla como segundo argumento, con la variable y los límites
> De igual forma que con derivadas, es posible crear integrales sin evaluar con la función `Integral`

```python
integrate(cos(x), x)
```

```python
integrate(cos(x), (x, 0, 1))
```

```python
integrate(exp(-x), (x, 0, oo))
```

```python
Integral(exp(-x ** 2 - y ** 2), (x, -oo, oo), (y, -oo, oo))
```

```python
integrate(exp(-x ** 2 - y ** 2), (x, -oo, oo), (y, -oo, oo))
```

```python
# cuando no puede calcular la integral, devuelve un objeto tipo Integral
integrate(x**x, x)
```

```python
integral = Integral((x**4 + x**2*exp(x) - x**2 - 2*x*exp(x) - 2*x -
    exp(x))*exp(x)/((x - 1)**2*(x + 1)**2*(exp(x) + 1)), x)
integral
```

```python
integral.doit()
```

```python
expr = Integral(log(x) ** 2, x)
expr.doit()
```

```python
Integral(sin(x ** 2), x).doit()
```

# Resolver sistemas de ecuaciones
Para solucionar sistemas de ecuaciones lineales, se utiliza la función `solve` o `solveset`

```python
ecuacion = Eq(x + 9, 8)
```

```python
solve(ecuacion)
```

```python
solveset(ecuacion)
```

```python
solve(Eq(4 * x + 4, x ** 2))
```

```python
solve(Eq(p + 1, x), p)
```

```python
solve(Eq(p + 1, x), x)
```

# Matrices
Para declarar matrices se utiliza la función `Matrix`

```python
A = Matrix([
    [2, 3, 4],
    [5, 2, -5],
    [3, -4, 5]
])
A
```

```python
# acceder a una fila
A.row(0)
```

```python
# acceder a una columna
A.col(2)
```

```python
A.T
```

```python
# sumar de forma individual, multiplicar, colocar dentro simbolos, multiplicar por un simbolo
B = Matrix([[3, 4, 5], [1, 1, 1], [0, 1, 2]])
A.multiply_elementwise(B)
```

```python
linsolve(Matrix(([1, 1], [2, 3])), (x, y))
```

```python
import numpy as np
A_np = np.array([
    [2, 3, 5],
    [3, 6, 2],
    [8, 3, 6]
])
b_np = np.array([52, 61, 75])
np.linalg.solve(A_np, b_np)
```

```python
A_2 = Matrix([[2, 3, 5], [3, 6, 2], [8, 3, 6]])
# x_2 = Matrix(3,1,[3,7,5])
b = Matrix(3, 1, [52, 61, 75])
soln = A_2.LUsolve(b)
soln
```

# Plots
