```python
x = np.arange(-10, 10.05, 0.05)
y = x ** 2
```

```python
# Colocar etiquetas y titulos
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfica de la función x ^ 2')
plt.show
```

```python
# Colocar grilla
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfica de la función x ^ 2')
plt.grid()
plt.show
```

```python
y_2 = - x ** 2 

plt.plot(x, y, label='x ^ 2')
plt.plot(x, y_2, label='- x ^ 2')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfica de la función x ^ 2')
plt.grid()
plt.legend()
plt.show
```

```python
# Agregar colores, estilos de linea, estilos de los puntos, grosor de la linea

plt.plot(x, y, label='x ^ 2', color='red', linestyle=':')
plt.plot(x, y_2, label='- x ^ 2', linewidth=5)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfica de la función x ^ 2')
plt.grid()
plt.legend()
# ajustar el grosor de la grafica
plt.tight_layout()
plt.show
```

# Gráficas de relleno
Utilizamos la función `fill_between`
> Debo pasarle dos argumentos: primero los datos del eje `x` y después los del eje `y`

```python
plt.plot(x, y_2, label='x ^ 2')
plt.fill_between(x, y_2, alpha=0.25)

plt.plot(x, y, label='x ^ 2')
plt.fill_between(x, y, alpha=0.25)

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfica de la función x ^ 2')
plt.legend()
# ajustar el grosor de la grafica
plt.tight_layout()

plt.show
```

# EJEMPLO REAL LIFE

```python
import pandas as pd
```

```python
dataframe = pd.read_csv('salaries.csv')
dataframe
```

```python
ages = dataframe['Age']
all_devs = dataframe['All_Devs']
python = dataframe['Python']
javascript = dataframe['JavaScript']
```

```python
plt.style.use('seaborn-notebook')

plt.plot(ages, all_devs, label='All Devs', linestyle="--", color="black")
# relleno azul cuando los salarios all_devs sea mayor a los salarios python
# pasar where a fill_between
plt.fill_between(ages, all_devs, python, where=(all_devs > python), interpolate=True, alpha=0.25)

plt.plot(ages, python, label='Python Devs', color='red')
plt.fill_between(ages, python, all_devs, where=(python > all_devs), interpolate=True, alpha=0.25)

#plt.plot(ages, javascript, label='JavaScript Devs')
#plt.fill_between(ages, javascript, all_devs, where=(javascript > all_devs), interpolate=True, alpha=0.25)

plt.title('Developer Salaries By Age')
plt.xlabel('Age')
plt.ylabel('Salary (USD)')
plt.tight_layout()
plt.legend()
plt.show
```

# DIAGRAMAS DE MOMENTO FLECTOR
> NOTA: En clase no me estaba cuadrando el diagrama de momento debido a que estaba tomando el eje `x` con un arange desde 0 hasta 4, debía ser de 0 hasta 10.005 ya que la longitud de la viga es 10 

```python
# en metros
LONGITUD = 10
# en kN
CARGA_P = 150
```

```python
x = np.arange(0, 10.005, 0.005)
# es mejor hacerlo así, usando los valores de la variable LONGITUD en lugar de escribir el valor a mano y equivocarse
# x = np.arange(0, LONGITUD + 0.005, 0.005)
len(x)
```

```python
y = np.full(len(x), CARGA_P / 2)
```

```python
plt.plot(x[x <= 5], y[x <= 5], label='V(x) = P / 2')
# es mejor así, usando la variable LONGITD
#plt.plot(x[x <= LONGITUD / 2], y[x <= LONGITUD / 2], label='V(x) = P / 2')
plt.fill_between(x[x <= 5], y[x <= 5], alpha=0.25, label='Cortante positivo')

plt.plot(x[x > 5], - y[x > 5], label='V(x) = - P / 2')
plt.fill_between(x[x > 5], -y[x > 5], alpha=0.25, label='Cortante negativo')

plt.legend()
plt.show()
```

```python
momento_tramo_1 = CARGA_P / 2 * x
momento_tramo_2 = CARGA_P / 2 * (LONGITUD - x)
```

```python
plt.plot(x[x <= 5], momento_tramo_1[x <= 5])
plt.plot(x[x >= 5], momento_tramo_2[x >= 5])

plt.fill_between(x[x <= 5], momento_tramo_1[x <= 5], alpha=0.25)
plt.fill_between(x[x >= 5], momento_tramo_2[x >= 5], alpha=0.25)
```
